# **DIAGRAMAS DE CAJAS  (BOXPLOTS)** 🍱

---

**¿QUE ES UN DIAGRAMA DE CAJA Y PARA QUE SIRVE?**
---

Es un diagrama que utiliza los los cinco estadísticos (mínimo, primer cuartil, mediana, tercer cuartil y máximo).

> * Mínimo: El valor más pequeño del conjunto de datos
* Primer cuartil: El valor medio de la mitad inferior de los datos
* Mediana: El valor medio del conjunto de datos
* Tercer cuartil: El valor medio de la mitad superior de los datos
* Máximo: El valor más grande del conjunto de datos

image.png

Los diagramas de cajas aparte de darte una manera visual y practica de entender la información tienen usos más especificos: identificar valores atipicos, mostrar la distribución de datos y comparar categorías.

---


**¿QUIEN INVENTO EL DIAGRAMA DE CAJAS?**
---

Fue **JOHN WILDER TUKEY** en 1977 (estadístico estadounidense). Su intención era poder entender los datos de una manera **VISUAL** antes de entrar en modelos y cosas estadisticas más complejas.

---

**¿Para qué sirven?**
---

Como ya escribi antes, los diagramas de cajas sirven para entender los datos de una forma mas visual:

* Muestran la dispersión y la distribución de un conjunto de datos.
* Sirven para identificar valores atípicos(outliers). Son los que se desvian más de lo normal.
* Sirven para comparar distribuciones entre grupos y categorias.
* Ya dan un resumen estadístico básico, mostrando los 5 estadisticos.


---

**DIFERENCIAS CON LOS DEMAS**
---
* La Diferencia con los historigramas es que son más compactos, no requieren tanta cosa como intervalos o bins, son mejores para comparar distribuciones lado a lado

* La Diferencia principal con los de tallo y hoja es que es un resumen general (varios grupos o categorias) y los de tallo y hoja son mas detallados (un grupo en especifico).

---

**¿Para qué son más usados?**
---

* Ciencias sociales: Comparar distribuciones de datos entre grupos. (por ejemplo, ingresos por género o región)

*	Medicina: Identificar rangos normales y extremos en mediciones como presión arterial o peso.

*	Finanzas: Analizar rendimientos de activos o fondos en diferentes períodos.

*	Educación: Comparar calificaciones o puntajes entre estudiantes de diferentes clases o exámenes.

*	Análisis de datos: Entender la distribución de datos antes de "moverle".

---

**Cómo hacer un diagrama de caja (boxplot)**
---
Organiza los datos: Ordena los datos numéricos de menor a mayor.


1. Calcula los valores clave:
   * Mediana
   * Primer cuartil
   * Tercer cuartil
   * Rango intercuartil

2. Determina los limites (bigotes)

3. Identifica valores atípicos: Cualquier dato fuera de los bigotes es un outlier.

4. Dibuja el gráfico

# **DIAGRAMA DE TALLO Y HOJA** 🌱
---

**¿Qué son?**
---

El diagrama de tallo y hoja también se conoce como Stem-and-Leaf Diagram, denominación que proviene del inglés.

Es muy similar al histograma, aunque este otro tipo de gráfico estadístico se utiliza para muestras de tamaño más grandes.

En términos simples, el diagrama de tallo y hojas es un esquema a través del cual se pueden graficar las veces que aparecen los valores de una determinada variable, es decir, la distribución de su frecuencia.

---

**¿Para que sirven los diagramas de tallo y hoja?**
---

Permite ordenar los datos de manera que se distingan incluso aquellos números repetidos entre las hojas un mismo tallo, lo cual ayuda a percibir asimetrías, secuencias e irregularidades en la variable objeto de estudio.

---

**¿Cómo se hacen los diagramas de tallo y hoja?**
---

**Para hacer un diagrama de tallo y hoja se deben seguir los siguientes pasos:**

* Ordenar los datos: organiza cada uno de los valores del conjunto numérico de menor a mayor.

* Agregar los tallos: van en una columna vertical, de menor a mayor y de arriba hacia abajo. Recuerda que los tallos son todos los números que se encuentran antes del último dígito, además, se deben colocar una sola vez y separados de las hojas mediante una línea.

* Colocar las hojas: de menor a mayor al lado de su tallo correspondiente y organizadas horizontalmente. Si se repite un valor entre las hojas, igualmente debes colocarlo.



---

**Ventajas y desventajas de usar estos diagramas**
---

**VENTAJAS**

* El diagrama de tallo y hoja permite ver la forma de una distribución.

* Se pueden comparar dos distribuciones usando el diagrama de tallo y hoja con doble tallo.

* Permite identificar rápidamente los valores atípicos de un conjunto de datos.

* Se puede determinar la moda de la serie de datos a simple vista.

**DESVENTAJAS**

* Cada vez se usa menos el diagrama de tallo y hoja, ya que un software informático puede hacer de manera rápida un gráfico más complejo.

* El tamaño de la muestra a representar es limitado, en general entre 15-150 datos.

* Solamente se pueden representar datos cuantitativos.

# **HISTORIGRAMAS** 📚
---

**¿QUE ES UN HISTORIGRAMA Y PARA QUE SIRVE?**
---

un histograma es un gráfico que, muestra la distribución de datos en una variable continua, se utiliza para analizar datos, identificar problemas y comparar distribuciones

**Caracteristicas:**

* Los histogramas se construyen a partir de intervalos de datos llamados bins

* Cada bin tiene una frecuencia, que es el número de datos en ese intervalo

* El eje horizontal muestra los valores de los datos

* El eje horizontal muestra los valores de los datos

---

**DIFERENCIAS CON LOS DEMAS**
---
* La diferencia fundamental entre histogramas y diagramas de barras es el tipo de datos que se representan. Los histogramas se usan en datos continuos, mientras que los diagramas de barras se usan en datos categóricos o nominales

* La diferencia con los diagramas de cajas es que los hitorigramas son un poco mas complejos y son utilizados para medir en intervalos cosa que los diagramas de caja no hacen.

---

**¿Cómo se crea un histograma?**
---

Se debe determinar el rango de valores de datos de cada barra (clases o intervalos).

Por ejemplo, al crear un histograma de edades en años, podríamos decidir clasificar por décadas (0-10, 11-20 y así). La altura de las barras nos mostraría la cantidad de personas en cada década

En el software, las clases se definen por el programa. Sin embargo, algunas herramientas de software permiten modificar el número de clases y sus puntos de inicio, lo que permite explorar y comprender mejor los datos

**Para construir un histograma, se deben seguir estos pasos:**


* Recopilar los datos numéricos

* Agrupar los datos en intervalos

* Contar la frecuencia de los datos en cada intervalo

* Dibujar los ejes

* Dibujar las barras

---

**TIPOS DE DATOS**
---

Los datos continuos son los mas apropiados para historigramas, ya que se miden sobre una escala con muchos valores posibles

edad

> * presión sanguínea
* peso
* temperatura
* velocidad

Si quieres usar datos nominales o categoricos es mejor usar un diagrama de barras puesto que se miden en una escala con solo algunos valores posibles.

En datos categóricos, la muestra suele dividirse en grupos, y las respuestas tienen un orden específico.
Por ejemplo, en una encuesta en la que se le pide su opinión en una escala de «muy en desacuerdo» a «muy de acuerdo» sus respuestas serán categórica
